```python
import csv
import matplotlib.pyplot as plt
f=open('seoul2.csv','r',encoding='cp949')
data=csv.reader(f)
next(data)
result=[]
result2=[]

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1]=='11' and row[0].split('-')[2] == '07':
            result.append(float(row[-1]))
    if row[-2] != '':
        if row[0].split('-')[1]=='11' and row[0].split('-')[2] == '07':
            result2.append(float(row[-2]))
        
 
plt.plot(result,color='red')
plt.plot(result2, color='blue')

plt.show()

```

```python
import csv
import matplotlib.pyplot as plt
f=open('seoul2.csv','r',encoding='cp949')
data=csv.reader(f)
next(data)
jan=[]
aug=[]
for row in data:
    month=row[0].split('-')[1]
    if row[-1]!='':
        if month=='01':
            jan.append(float(row[-1]))
        month=row[0].split('-')[1]
    if row[-1]!='':
        if month=='08':
            aug.append(float(row[-1]))
            
plt.hist(jan, bins=100, color='b',label='jan')            
plt.hist(aug, bins=100, color='r', label='aug')
plt.legend()
plt.show()
```
