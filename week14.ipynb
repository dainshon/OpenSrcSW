```python
import csv
import matplotlib.pyplot as plt
import numpy as np
f=open('seoul3.csv','r',encoding='cp949')
data=csv.reader(f)
next(data)
pop=[]
min=-1 #표준편차
max=-1
min_num=0 #표준편차 작은 row
max_num=0
total=0
excel=[]

for row in data:
    excel.append(row)
    pop.append(row[3:])           

f.close()

excel=np.array(excel)
a=np.array(pop)
y=np.array(a[5])

for i in range(101):
    y[i]=y[i].replace(",","")

y=y.astype(int) #타입변환
x=np.linspace(0,100,101)

for i in range(25): #25번돌면됨
    total=0
    row=np.array(a[i+1])

    for j in range(101):
        row[j]=row[j].replace(",","")
    row=row.astype(int)
    for j in range(101):
        total=total+pow(y[j]-row[j],2)
        
    if min<0 or total<min:
        if i+1 != 5:
            min=total
            y1=row
            min_num=i+1

    if total>max:
        max=total
        y2=row
        max_num=i+1
print(excel[min_num][0])
print(excel[max_num][0])
print(y1)
print(y2)
plt.style.use('ggplot')
plt.plot(x,y, color='red', label=excel[5][0])
plt.plot(x,y1,color='blue',label=excel[min_num][0]+'표준편차'+str(min))
plt.plot(x,y2,color='yellow',label=excel[max_num][0]+'표준편차: '+str(max))
plt.legend()
plt.show()
```
