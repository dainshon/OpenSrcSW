```python
import csv
f=open('seoul.csv','r',encoding='cp949')
data=csv.reader(f)
header=next(data)
max_temprange=0
temprange_date=''
min_temp=0
min_date=''
max_temp=0
max_date=''

for row in data:
    temprange=float(row[4])-float(row[3])
    if temprange>max_temprange:
        max_temprange=temprange
        temprange_date=row[0]
    if float(row[3])<min_temp:
        min_temp=float(row[3])
        min_date=row[0]
    if float(row[4])>max_temp:
        max_temp=float(row[4])
        max_date=row[0]
        
print("일교차가 가장 큰 날")
print(temprange_date)
print("최저기온이 가장 낮은 날")
print(min_date)
print("최고기온이 가장 높은 날")
print(max_date)
    
f.close()
```

```python
import csv
f=open('subwayfee.csv',encoding='UTF8')
data=csv.reader(f)
next(data)
subway='1호선'

max_p=0
min_p=0
max=[0]*3
max_station=['']*3
min=[999999]*3
min_station=['']*3

for row in data:
    if row[1]==subway:
        row[4]=int(row[4].replace(",",""))
        row[5]=int(row[5].replace(",",""))
        people=row[4]+row[5]
        #print(row[3])
        
        if people>max[0]:
            max[2]=max[1]
            max[1]=max[0]
            max[0]=people
            max_station[2]=max_station[1]
            max_station[1]=max_station[0]
            max_station[0]=row[3]
        elif people<=max[0] and people>max[1]:
            max[2]=max[1]
            max[1]=people
            max_station[2]=max_station[1]
            max_station[1]=row[3]
        elif people<=max[1] and people>max[2]:
            max[2]=people
            max_station[2]=row[3]
        
        if people<min[0]:
            min[2]=min[1]
            min[1]=min[0]
            min[0]=people
            min_station[2]=min_station[1]
            min_station[1]=min_station[0]
            min_station[0]=row[3]
        elif people>=min[0] and people<min[1]:
            min[2]=min[1]
            min[1]=people
            min_station[2]=min_station[1]
            min_station[1]=row[3]
        elif people>=min[1] and people<min[2]:
            min[2]=people
            min_station[2]=row[3]
            
    else:
        print("\n",subway)
        print("이용객이 가장 많은 역 순위")
        for i in range(3):
            print((i+1),". ",max_station[i],"(",max[i],")")
        print("\n이용객이 가장 적은 역 순위")
        for i in range(3):
            print((i+1),". ",min_station[i],"(",min[i],")")
        subway=row[1]
        max_p=0
        min_p=0
        max=[0]*3
        max_station=['']*3
        min=[999999]*3
        min_station=['']*3
f.close()
```
